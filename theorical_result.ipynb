```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd

data = {
    "AI model": ["Model A", "Model A", "Model A", "Model B", "Model B", "Model B"] * 3,
    "Technique": ["Technique 1"] * 6 + ["Technique 2"] * 6 + ["Technique 3"] * 6,
    "Data Used": ["Dataset 1", "Dataset 2", "Dataset 3"] * 6,
    "F1 Score": [0.83, 0.85, 0.89, 0.76, 0.80, 0.83] + [0.87, 0.89, 0.91, 0.83, 0.85, 0.88] + [0.9, 0.95, 0.97, 0.84, 0.87, 0.91],
    "Computational Time": [120, 125, 130, 95, 100, 105] + [130, 135, 140, 105, 115, 125] + [155, 165, 170, 135, 145, 155],
    "Memory Used": [250, 260, 270, 200, 210, 220] + [280, 270, 320, 250, 250, 220] + [310, 340, 380, 310, 320, 310],
}

df = pd.DataFrame(data)
```

```python
fig, axes = plt.subplots(3, 1, figsize=(10, 15))

# Plot F1 Score
sns.barplot(data=df, x="AI model", y="F1 Score", hue="Technique", ci=None, ax=axes[0])
axes[0].set_xlabel("AI Model")
axes[0].set_ylabel("F1 Score")
axes[0].set_title("F1 Score Comparison")


# Plot Computational Time
sns.barplot(data=df, x="AI model", y="Computational Time", hue="Technique", ci=None, ax=axes[1])
axes[1].set_xlabel("AI Model")
axes[1].set_ylabel("Computational Time")
axes[1].set_title("Computational Time Comparison")


# Plot Memory Usage
sns.barplot(data=df, x="AI model", y="Memory Used", hue="Technique", ci=None, ax=axes[2])
axes[2].set_xlabel("AI Model")
axes[2].set_ylabel("Memory Used")
axes[2].set_title("Memory Usage Comparison")


# Adjust layout to prevent overlapping
plt.tight_layout()

plt.show()
```
